```python
new_data=data.drop(["RowNumber","CustomerId","Surname","Geography","Gender"],axis='columns')
```

```python
new_data.fillna(method='ffill',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
rux= df.drop('Exited', axis='columns') 
ruy= df['Exited']
```

```python
ru = RandomUnderSampler(random_state=42)
x, y = ru.fit_resample(rux, ruy)
```

```python
# x['Balance'].value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
x_train_scaled = scaler.fit_transform(x_train)

# Transform the test data
x_test_scaled = scaler.transform(x_test)
```

```python
rfc=RandomForestClassifier(max_depth=6, min_samples_leaf=5, min_samples_split=10)
rfc.fit(x_train_scaled,y_train)
```

```python
rfc.score(x_test_scaled,y_test)
```

```python
y_predict = rfc.predict(x_test_scaled)
y_pred=rfc.predict(x_train_scaled)
```

```python
test_a = accuracy_score(y_test,y_predict)
train_a = accuracy_score(y_train,y_pred)
print("test_accuracy",test_a)
print("train_accurary",train_a)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
precision = precision_score(y_test, y_predict)
recall = recall_score(y_test, y_predict)
print("recall:",recall)
print("precision:", precision)
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_predict)
f1
```
